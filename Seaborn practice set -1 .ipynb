```python
# Python program to illustrate 
# Plotting categorical scatter 
# plots with Seaborn 

# importing the required module 
import matplotlib.pyplot as plt 
import seaborn as sns 

# x axis values 
x =['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu'] 

# y axis values 
y =[5, 6.7, 4, 6, 2, 4.9, 1.8] 

# plotting strip plot with seaborn 
ax = sns.stripplot(x = x, y = y); 

# giving labels to x-axis and y-axis 
ax.set(xlabel ='Days', ylabel ='Amount_spend') 

# giving title to the plot 
plt.title('My first graph'); 

# function to show plot 
plt.show() 
```

```python
df = sns.load_dataset('iris')

df.head()
```

```python
sns.set_context('paper')

sns.stripplot(x = 'species', y = 'sepal_length', data = df, hue = 'species', palette = 'deep')
```

```python
sns.set_context('paper')

sns.stripplot(x = 'species', y = 'sepal_length', data = df, hue = 'species', palette = 'dark')
```

```python
sns.set_context('paper')
sns.set_style('darkgrid')

sns.stripplot(x = 'species', y = 'sepal_length', data = df, hue = 'species', palette = 'muted')
```

```python
sns.set_context('paper')

sns.stripplot(x = 'species', y = 'sepal_length', data = df, hue = 'species', palette = 'dark', dodge = True)
plt.xlabel('Species', fontdict= {'size':13, 'weight':'bold'}, labelpad = 12)
plt.ylabel('Sepal length', fontdict= {'size':13, 'weight':'bold'}, labelpad = 12)
```

```python
sns.set_style('whitegrid')
sns.swarmplot(x = 'sepal_length', y = 'species', data = df, hue = 'species', palette = 'dark', dodge = True)
plt.xlabel('Species', fontdict= {'size':13, 'weight':'bold'}, labelpad = 12)
plt.ylabel('Sepal length', fontdict= {'size':13, 'weight':'bold'}, labelpad = 12)
```

```python
df = sns.load_dataset('tips')

df.head()
```

```python
df.info()
```

```python
cust_plot = sns.FacetGrid(df, row = 'smoker', col = 'sex', hue = 'day', palette = 'bright')

cust_plot.map(plt.bar, 'total_bill', 'tip').add_legend()
plt.tight_layout()
plt.show()
```
